# Set up plotting. 

```python
def plot_precision_recall(filename, datasets, ylim=[0, 5], 
                          dotted=False, rotate_xlabels=False, legend_labels=None, legend_colors=None, xlabel=None, 
                          figsize=(10, 9), log=True):
    fig, subs = plt.subplots(3, 2, sharey=True, figsize=figsize)    
    
    for k, gen in enumerate(['1', '2']):
        for l, attr in enumerate(['precision', 'recall', 'F1']):
            print(attr + gen + '_estimates')
            error_estimates = [d[attr + gen + '_estimates'] for d in datasets]
            upper_bounds = [d[attr + gen + '_upper_bounds'] for d in datasets]
            
            upper_bound = [np.median(ub) for ub in upper_bounds]
            lower_bound = [0]*len(datasets)
            for j in range(len(datasets)):
                if log:
                    subs[l, k].fill([j+0.7, j+0.7, j+1.3, j+1.3], [-np.log10(1-lower_bound[j]), -np.log10(1-upper_bound[j]), -np.log10(1-upper_bound[j]), -np.log10(1-lower_bound[j])],
                                lw=0, c='lightgray', zorder=-10)
                else:
                    subs[l, k].fill([j+0.7, j+0.7, j+1.3, j+1.3], [lower_bound[j], upper_bound[j], upper_bound[j], lower_bound[j]],
                                lw=0, c='lightgray', zorder=-10)

            if log:
                data = [-np.log10(1-np.clip(ee, None, ub)) for ee, ub in zip(error_estimates, upper_bounds)]
            else:
                data = [np.clip(ee, None, ub) for ee, ub in zip(error_estimates, upper_bounds)]
            print('n=', [d.shape[0] for d in data])
            print(attr+gen, [1-np.median(x) for x in error_estimates])
            parts = subs[l, k].violinplot(data, showmedians=True, showextrema=False, widths=0.8)


            if dotted:
                if log:
                    subs[l, k].plot(np.arange(1, 1+len(datasets)), 
                                [np.median(-np.log10(1-datasets[0][attr + gen + '_estimates']))]*len(datasets), 
                                color='black', linestyle='--')
                else:
                    subs[l, k].plot(np.arange(1, 1+len(datasets)), 
                                [np.median(datasets[0][attr + gen + '_estimates'])]*len(datasets), 
                                color='black', linestyle='--')


            for pc, d in zip(parts['bodies'], datasets):
                pc.set_facecolor(d['color'])
                pc.set_edgecolor('black')
                pc.set_alpha(1)
            parts['cmedians'].set_edgecolor('black')
            parts['cmedians'].set_linewidth(2)
                
            subs[l, k].grid(True, axis='y', color='darkgray')
            subs[l, k].set_axisbelow(True)
            subs[l, k].set_ylim(ylim)

    # yticks only on leftmost panels
    if log:
        yticks = np.arange(1, ylim[1])
        ytick_labels = ['0.' + '9'*int(f) % f for f in yticks]
    else:
        yticks = np.arange(0.9, 1.02, 0.2)
        ytick_labels = [str(x) for x in yticks]
    subs[0, 0].set_yticks(yticks)
    subs[0, 0].set_yticklabels(ytick_labels, {'fontsize': 15})
    subs[0, 1].tick_params('y', length=0)
    subs[1, 0].set_yticks(yticks)
    subs[1, 0].set_yticklabels(ytick_labels, {'fontsize': 15})
    subs[1, 1].tick_params('y', length=0)
    subs[2, 0].set_yticks(yticks)
    subs[2, 0].set_yticklabels(ytick_labels, {'fontsize': 15})
    subs[2, 1].tick_params('y', length=0)
    subs[0, 0].set_ylabel('Precision', {'fontsize':18})
    subs[1, 0].set_ylabel('Recall', {'fontsize':18})
    subs[2, 0].set_ylabel('$F_1$ Score', {'fontsize':18})
    
    # xticks only on bottom panels
    subs[0, 0].set_xticks([])
    subs[0, 1].set_xticks([])
    subs[1, 0].set_xticks([])
    subs[1, 1].set_xticks([])
    subs[2, 0].set_xticks(np.arange(1, 1+len(datasets)))
    subs[2, 0].set_xticklabels([d['name'] for d in datasets], {'fontsize': 15})
    subs[2, 1].set_xticks(np.arange(1, 1+len(datasets)))
    subs[2, 1].set_xticklabels([d['name'] for d in datasets], {'fontsize': 15})
    if rotate_xlabels:
        for tick in subs[2, 0].get_xticklabels():
            tick.set_rotation(90)
        for tick in subs[2, 1].get_xticklabels():
            tick.set_rotation(90)
    if xlabel is not None:
        fig.text(0.5, 0.1, xlabel, ha='center', va='center', fontsize=15)
    
    # titles only on topmost panels
    subs[0, 0].set_title('Heterozygous Sites', {'fontsize': 18}, loc='left')
    subs[0, 1].set_title('Homozygous Alt. Sites', {'fontsize': 18}, loc='left')
    subs[1, 0].set_title(None)
    subs[1, 1].set_title(None)
    subs[2, 0].set_title(None)
    subs[2, 1].set_title(None)
    
    # legend
    if legend_labels is not None:
        entries = []
        for label, color in zip(legend_labels, legend_colors):
            entries.append(patches.Patch(color=color, label=label))
        subs[2, 1].legend(handles=entries, loc='lower center', ncol=3,
                                    bbox_to_anchor=(0, -0.5), borderaxespad=0., fontsize=15, frameon=False)
    
    
    fig.subplots_adjust(wspace=0.0, hspace=0.0, top=0.9, bottom=0.25 if rotate_xlabels else 0.2)
    #plt.show()
    if filename is not None:
        plt.savefig(filename)
    else:
        plt.show()
```

```python

datasets_of_interest = [
    {'name': 'ihart v3.2',
     'color': '#009668',
     'family_genotype_dir': '../../DATA/ihart.v32/family_genotype_counts',
     'param_file': '../../FamilySeqError/params/ihart.v32_params.json',
     'sample_filter': lambda x: x.is_child
    },
    
    {'name': 'ihart v3.4',
     'color': '#cedd38',
     'family_genotype_dir': '../../DATA/ihart.v34/family_genotype_counts',
     'param_file': '../../FamilySeqError/params/ihart.v34_params.json',
     'sample_filter': lambda x: x.is_child
    }
]


compare_rates.process_datasets(datasets_of_interest, chroms)
```

```python
def plot_comparison(dataset1, dataset2, label1, label2, figname, log=False):
    fig = plt.figure(figsize=(11, 4))

    for i, (attr, title) in enumerate([('precision', 'Precision'), 
                                       ('recall', 'Recall'), 
                                       ('F1', '$F_1$ Score')]):
        ax = plt.subplot(1, 3, i+1)

        shared_samples = sorted(set(dataset1['%s1_samples' % attr]) & set(dataset2['%s1_samples' % attr]))
        print(len(shared_samples))
        data = np.zeros((len(shared_samples), 2))

        sample_to_index = dict([(x, i) for i, x in enumerate(dataset1['%s1_samples' % attr])])
        for k, s in enumerate(shared_samples):
            if log:
                data[k, 0] = -np.log10(1-dataset1['%s1_estimates' % attr][sample_to_index[s]])
            else:
                data[k, 0] = dataset1['%s1_estimates' % attr][sample_to_index[s]]
        sample_to_index = dict([(x, i) for i, x in enumerate(dataset2['%s1_samples' % attr])])
        for k, s in enumerate(shared_samples):
            if log:
                data[k, 1] = -np.log10(1-dataset2['%s1_estimates' % attr][sample_to_index[s]])
            else:
                data[k, 1] = dataset2['%s1_estimates' % attr][sample_to_index[s]]

        plt.scatter(data[:, 0], data[:, 1], color='#4db6ac', label='Heterozygous Sites')
        if log:
            fold_change = (10**-data[:, 1])/(10**-data[:, 0])
        else:
            fold_change = (1-data[:, 1])/(1-data[:, 0])
        print('Med %s het fold change' % title, np.median(fold_change), np.median(1/fold_change))
        shared_samples = sorted(set(dataset1['%s2_samples' % attr]) & set(dataset2['%s2_samples' % attr]))
        data = np.zeros((len(shared_samples), 2))

        sample_to_index = dict([(x, i) for i, x in enumerate(dataset1['%s2_samples' % attr])])
        for k, s in enumerate(shared_samples):
            if log:
                data[k, 0] = -np.log10(1-dataset1['%s2_estimates' % attr][sample_to_index[s]])
            else:
                data[k, 0] = dataset1['%s2_estimates' % attr][sample_to_index[s]]
        sample_to_index = dict([(x, i) for i, x in enumerate(dataset2['%s2_samples' % attr])])
        for k, s in enumerate(shared_samples):
            if log:
                data[k, 1] = -np.log10(1-dataset2['%s2_estimates' % attr][sample_to_index[s]])
            else:
                data[k, 1] = dataset2['%s2_estimates' % attr][sample_to_index[s]]
        plt.scatter(data[:, 0], data[:, 1], color='#ce93d8', label='Homozygous Alt. Sites')
        if log:
            fold_change = (10**-data[:, 1])/(10**-data[:, 0])
        else:
            fold_change = (1-data[:, 1])/(1-data[:, 0])
        print('Med %s homalt fold change' % title, np.median(fold_change), np.median(1/fold_change))
        if i==0:
            plt.ylabel(label2, fontsize=18)
        elif i==1:
            plt.xlabel(label1, fontsize=18)
            
        if log:
            ticks = [1, 2, 3, 4, 5]
            ticklabels = [str(1-(10**-x)) for x in ticks]
        else:
            ticks = [0.985, 0.990, 0.995, 1.0]
            ticklabels = [str(x) for x in ticks]
        if i==0:
            plt.yticks(ticks, ticklabels, fontsize=15)
        else:
            plt.yticks(ticks, ['' for x in ticks], fontsize=15)
        plt.xticks(ticks[:-1], ticklabels[:-1], fontsize=15)
        plt.plot(ticks, ticks, color='black', linestyle='--')
        plt.xlim([ticks[0], ticks[-1]])
        plt.ylim([ticks[0], ticks[-1]])
        plt.title(title, fontsize=18)
        #plt.legend(fontsize=14)
        plt.grid(True, linewidth=2, zorder=0, color='lightgrey', linestyle=':')
        ax.set_axisbelow(True)

        if i==0:
            plt.legend(loc='lower center', ncol=2,
                   bbox_to_anchor=(1.5, -0.45), borderaxespad=0., fontsize=14, frameon=True)

    fig.subplots_adjust(wspace=0.05, hspace=0.0, top=0.9, bottom=0.3)

    #plt.tight_layout()
    plt.savefig(figname)
    plt.show()
```

```python
plot_comparison(datasets_of_interest[0], datasets_of_interest[1], 'GATK v3.2', 'GATK v3.4', '../plots/gatk.eps')
```

```python
num_sites = [compare_rates.pull_num_sites(d, chroms) for d in datasets_of_interest]
print([np.nanmedian(x) for x in num_sites])

plt.violinplot([np.log10(x[~np.isnan(x)]) for x in num_sites], showmedians=True, showextrema=False, widths=0.8)
plt.show()
```

```python
fig = plt.figure(figsize=(11, 5))

def plot_comparison(d, attr, title, marker, markersize=None, label=None):

    lcl_samples = [x for x in d['%s1_samples' % attr] if x.endswith('_LCL') and x[:-4] in d['%s1_samples' % attr]]
    wb_samples = [x[:-4] for x in lcl_samples]
    print(len(lcl_samples))

    data = np.zeros((len(lcl_samples), 2))
    sample_to_index = dict([(x, i) for i, x in enumerate(d['%s1_samples' % attr])])
    for k, (lcl_s, wb_s) in enumerate(zip(lcl_samples, wb_samples)):
        data[k, 0] = d['%s1_estimates' % attr][sample_to_index[wb_s]]
        data[k, 1] = d['%s1_estimates' % attr][sample_to_index[lcl_s]]
    plt.plot(-np.log10(1-data[:, 0]), -np.log10(1-data[:, 1]), color='#4db6ac', marker=marker, markersize=markersize, linewidth=0, label=None if label is None else 'Heterozygous Sites in %s' % label)
    print('%s Het' % title, np.mean(data[:, 1]-data[:, 0]))
    print(stats.ttest_rel(data[:, 0], data[:, 1]))
    print(np.median(data[:, 0]), np.median(data[:, 1]))
    print(data.shape, stats.pearsonr(data[:, 0], data[:, 1]))
    lcl_samples = [x for x in d['%s1_samples' % attr] if x.endswith('_LCL') and x[:-4] in d['%s2_samples' % attr]]
    wb_samples = [x[:-4] for x in lcl_samples]
    print(len(lcl_samples))

    data = np.zeros((len(lcl_samples), 2))
    sample_to_index = dict([(x, i) for i, x in enumerate(d['%s2_samples' % attr])])
    for k, (lcl_s, wb_s) in enumerate(zip(lcl_samples, wb_samples)):
        data[k, 0] = d['%s2_estimates' % attr][sample_to_index[wb_s]]
        data[k, 1] = d['%s2_estimates' % attr][sample_to_index[lcl_s]]
    plt.plot(-np.log10(1-data[:, 0]), -np.log10(1-data[:, 1]), color='#ce93d8', marker=marker, markersize=markersize, linewidth=0, label=None if label is None else 'Homozygous Alt. Sites in %s' % label)
    print('%s Homalt' % title, np.mean(data[:, 1]-data[:, 0]))
    print(stats.ttest_rel(data[:, 0], data[:, 1]))
    print(np.median(data[:, 0]), np.median(data[:, 1]))
    print(stats.pearsonr(data[:, 0], data[:, 1]))

    #ticks = [0.996, 0.997, 0.998, 0.999, 1.0]
    ticks = [1, 2, 3, 4]
    plt.xticks(ticks[:-1], [str(1-10**-x) for x in ticks[:-1]], fontsize=15)
    plt.plot(ticks, ticks, color='black', linestyle='--')
    plt.xlim([ticks[0]-0, ticks[-1]])
    plt.ylim([ticks[0], ticks[-1]])
    plt.title(title, fontsize=18)
    plt.grid(True, linewidth=2, zorder=0, color='lightgrey', linestyle=':')

plt.subplot(1, 3, 1)
plot_comparison(datasets_of_interest[0], 'precision', 'Precision', 'o')
plot_comparison(datasets_of_interest[1], 'precision', 'Precision', '*', 12, label='LCR')
plt.legend(loc='lower center', ncol=1,
           bbox_to_anchor=(0.8, -0.45), borderaxespad=0., fontsize=14, frameon=True)
plt.ylabel('Lymphoblastoid Cell Line', fontsize=18)
plt.yticks([1, 2, 3, 4], [str(1-10**-x) for x in [1, 2, 3, 4]], fontsize=15)

plt.subplot(1, 3, 2)
plot_comparison(datasets_of_interest[0], 'recall', 'Recall', 'o')
plot_comparison(datasets_of_interest[1], 'recall', 'Recall', '*', 12)
plt.xlabel('Whole Blood', fontsize=18)
plt.yticks([1, 2, 3, 4], ['' for x in [1, 2, 3, 4]], fontsize=15)

plt.subplot(1, 3, 3)
plot_comparison(datasets_of_interest[0], 'F1', '$F_1$ Score', 'o', label='HCR')
plot_comparison(datasets_of_interest[1], 'F1', '$F_1$ Score', '*', 12)
plt.legend(loc='lower center', ncol=1,
           bbox_to_anchor=(0.1, -0.45), borderaxespad=0., fontsize=14, frameon=True)
plt.yticks([1, 2, 3, 4], ['' for x in [1, 2, 3, 4]], fontsize=15)

#plt.tight_layout()

fig.subplots_adjust(wspace=0.05, hspace=0.0, top=0.9, bottom=0.3)

plt.savefig('../plots/lcl.eps')
plt.show()
```
